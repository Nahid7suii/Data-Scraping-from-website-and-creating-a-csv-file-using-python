```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_manufacturers_by_motor_vehicle_production#2016'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find('table')
```

```python
soup.find_all('table')[5]
```

```python
table = soup.find_all('table')[3]
print(table)
```

```python
table_title = table.find_all('th')
table_title
```

```python
car_table_title = [title.text.strip() for title in table_title]
print(car_table_title)
```

```python
df = pd.DataFrame(columns = car_table_title)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
    
```

```python
df
```

```python
df.to_csv('D:\Data Analysis\Data\scraped_data_project.csv', index=False)
print('Table has been Scraped and Scraped_data_project.csv has been created')
```
